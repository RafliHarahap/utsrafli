```python
# Data yang diberikan
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

# Fungsi untuk mengembalikan perusahaan beserta jumlah karyawan berdasarkan kebangsaannya
def get_companies_with_employee_nationality(employee_list, company_list):
    # Membuat kamus untuk perusahaan dan kebangsaannya
    company_nationalities = {}
    
    # Iterasi melalui daftar perusahaan untuk inisialisasi perusahaan
    for company in company_list:
        company_nationalities[company['name']] = {}

    # Iterasi melalui daftar karyawan untuk menghitung kebangsaan di setiap perusahaan
    for employee in employee_list:
        company_name = employee['company']
        nationality = employee['nationality']
        
        # Jika kebangsaan sudah ada di perusahaan, tambahkan jumlahnya
        if nationality in company_nationalities[company_name]:
            company_nationalities[company_name][nationality] += 1
        else:
            company_nationalities[company_name][nationality] = 1

    # Mengubah hasil ke format yang diinginkan
    result = []
    for company in company_list:
        result.append({
            "company": company['name'],
            "employee_nationality": company_nationalities[company['name']]
        })
    
    return result

# Panggil fungsi dan tampilkan hasilnya
companies_with_nationality = get_companies_with_employee_nationality(employee_detail_list, company_detail_list)
for company in companies_with_nationality:
    print(company)
```
